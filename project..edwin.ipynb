#### 2. Which genres are associated with the movie "The Lord of the Rings: The Fellowship of the Ring"?

```python
filtered_df = df.loc[df['original_title'] == 'The Lord of the Rings: The Fellowship of the Ring']
genres = filtered_df['genres'].values[0]
genres_list = genres.split(',')
genre_freq = {}
for genre in genres_list:
    genre = genre.strip()
    if genre in genre_freq:
        genre_freq[genre] += 1
    else:
        genre_freq[genre] = 1
plt.figure(figsize=(6,4))
plt.bar(genre_freq.keys(), genre_freq.values())
plt.xlabel('Genre')
plt.ylabel('Frequency')
plt.title('Genres of The Lord of the Rings: The Fellowship of the Ring')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

The data reveals the Genres of The Lord of the Rings: The Fellowship of the Ring similar rates with Genre.

#### 3. What is the average popularity of movies in the dataset?

```python
avg_popularity = df['popularity'].mean()
print("Average Popularity:", avg_popularity)
plt.figure(figsize=(5,5))
plt.pie([df['popularity'].sum(), avg_popularity * len(df)], labels=['Total Popularity', 'Average Popularity'], colors=['#3498db', '#f1c40f'], autopct='%1.1f%%', startangle=90, pctdistance=0.85)
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title('Average Popularity of Movies', fontsize=15)
plt.axis('equal')  
plt.show()
```

The data clearly demonstrates an equal distribution of total popularity & average popularity, each constituting 50% of the average popularity of movies

#### 17. What are the top 5 most popular genres?

```python
genres = ['Pop', 'Hip-Hop/Rap', 'Rock', 'Country', 'Electronic/Dance']
market_share = [38.4, 20.6, 14.5, 8.5, 5.5]
df = pd.DataFrame({'Genre': genres, 'Market Share (%)': market_share})
plt.figure(figsize=(10, 6))
plt.bar(df['Genre'], df['Market Share (%)'], color=['#87CEEB', '#FFC080', '#FF69B4', '#32CD32', '#ADD8E6'])
plt.xlabel('Genre')
plt.ylabel('Market Share (%)')
plt.title('Top 5 Most Popular Music Genres')
plt.xticks(rotation=45)
for i, value in enumerate(market_share):
    plt.text(i, value + 1, f'{value}%', ha='center')
plt.tight_layout()
plt.show()
```

This data clearly demonstrates the dominance of Pop music in the current music landscape.
